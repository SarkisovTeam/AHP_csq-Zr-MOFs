```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data from CSV
data = pd.read_csv('mech_prop.csv')

# Set style
sns.set_style("whitegrid")

# Plotting
plt.figure(figsize=(18, 6))

# Example plot (you can choose the type of plot suitable for your data)
ax = sns.barplot(data=data, x='MOFs', y='Youngs_modulus_Hill_average_GPa', palette='viridis')

# Highlight MOF-399
highlight_color = 'red'
highlight_value_mof399 = data[data['MOFs'] == 'MOF-399']['Youngs_modulus_Hill_average_GPa'].values[0]
highlight_value_dut6 = data[data['MOFs'] == 'DUT-6']['Youngs_modulus_Hill_average_GPa'].values[0]
highlight_value_dut60 = data[data['MOFs'] == 'DUT-60']['Youngs_modulus_Hill_average_GPa'].values[0]
ax.axhline(highlight_value_mof399, color='red', linestyle='--')

# Customize plot
plt.title('', fontsize=16, fontweight='bold')
plt.xlabel('', fontsize=16, fontweight='bold')
plt.ylabel('Young’s modulus (GPa)', fontsize=24)
plt.xticks(rotation=45, ha='right', fontsize=24)
plt.yticks(fontsize=24)
plt.ylim(0, None)  # Ensure y-axis starts from 0

# Customizing color for MOF-399, DUT-6, and DUT-60
for bar in ax.patches:
    if bar.get_height() == highlight_value_mof399:
        bar.set_color('red')
    elif bar.get_height() == highlight_value_dut6:
        bar.set_color('pink')
    elif bar.get_height() == highlight_value_dut60:
        bar.set_color('orange')
    bar.set_edgecolor('black')  # Add black border around each bar

# Increase font size for specific x-axis labels
x_labels = ax.get_xticklabels()
for label in x_labels:
    if label.get_text() in ['T', 'TT', 'TTT', 'TBT', 'TBT-N', 'BB', 'BBB', '5T', 'MOF-399', 'DUT-6', 'DUT-60']:
        label.set_fontsize(24)

# Add grid lines behind bars
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Remove spines
#sns.despine()

plt.tight_layout()

# Save the plot
plt.savefig('Young_modulus_Hill_average_GPa_3B_13.04.2024.png', dpi=600)

# Show plot (optional)
plt.show()
```
